```python
def check_seed(input_window, threshold):
    (x,y) = input_window.shape
    I_c = input_window[int(x/2), int(y/2)]
    I_avg = np.sum(input_window)/(x*y)
    return True if I_c<=threshold*I_avg else False
    
wx = 28
wy = 28
threshold = 0.9
output = np.zeros(ori_img.shape)
for i in range(ori_img.shape[0]-wx):
    for j in range(ori_img.shape[1]-wy):
        output[i,j] = check_seed(ori_img[i:i+wx, j:j+wy], threshold)
        
```
