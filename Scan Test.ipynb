```python
from helper import rolling_window

img = cv2.imread("C:/Users/Dominic/Desktop/wallpaper.jpg", 0).astype(np.float32)
print(img.shape)

stride = (100, 100)
window_size = (227, 227)
imgs = rolling_window(img, window_size, stride)
print(imgs.shape)

plt.figure(figsize=(10,5))
plt.subplot(121),plt.imshow(img, cmap='gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.show()


nx, ny = imgs.shape[0], imgs.shape[1]
plt.figure(figsize=(20, 20 * nx / ny))

for i in range(nx):
    for j in range(ny):
        plt.subplot(nx, ny, i * ny + j + 1)
        plt.imshow(imgs[i, j], cmap='gray')
        plt.axis('off')
        
plt.subplots_adjust(wspace=1, hspace=0)
plt.show()
```
