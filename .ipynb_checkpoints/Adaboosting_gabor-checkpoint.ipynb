```python
def normalize_grayscale(image, beta=0.75, image_id_k = 1, A_kp=None):
    N,M = image.shape
    k = image_id_k
    
    if image_id_k > 1 and  A_kp is None :
        print("please define A_kp as this is not the first image")
    elif image_id_k == 1:
        A_kp = np.zeros((N,M))
        
    alpha = beta*(k-1)/k
    a = np.mean(image, axis=0)
    A_k = alpha*A_kp + (1-alpha)*np.tile(a,(N,1))
    image = 128*(image/A_k)
    return image, A_k
```

```python

def build_filters():
    filters = []
    ksize = 31
    for theta in np.arange(0, np.pi, np.pi / 16):
        kern = cv2.getGaborKernel((ksize, ksize), 4.0, theta, 10.0, 0.5, 0, ktype=cv2.CV_32F)
        kern /= 1.5*kern.sum()
        filters.append(kern)
    return filters
```

```python
def process(img, filters):
    accum = np.zeros_like(img)
    for kern in filters:
        fimg = cv2.filter2D(img, cv2.CV_8UC3, kern)
        np.maximum(accum, fimg, accum)
    return accum
```

```python
def naive_gabor(ksize=31, sigma=1, theta=np.pi/8, lamb=10, gamma=0.5, psi=0, ktype=cv2.CV_32F):
    kern = cv2.getGaborKernel((ksize, ksize), 4.0, theta, 10.0, 0.5, 0, ktype=cv2.CV_32F)
    kern /= 1.5*kern.sum()
    return kern
```

```python
def f_j(img, kern, ddepth=cv2.CV_8UC3):
    fimg = cv2.filter2D(img, ddepth, kern)
    print('f_j, shape is {}'.format(fimg.shape))
    return fimg
```

```python
def normalize(v):
    norm=np.linalg.norm(v, ord=1)
    if norm==0:
        norm=np.finfo(v.dtype).eps
    return v/norm
```

```python
def train_weak_classifier_for_certain_filter(kern, size=200):
    reg = linear_model.LinearRegression()
    A_kp=None
    X=np.zeros([size*2,227*227])
    for i in range(1,size+1):
        img_fn = '../project/Positive/'+("%05d" % i) +'.jpg'
        img = cv2.imread(img_fn,0)
        img, A_kp = normalize_grayscale(img, beta=0.75, image_id_k = i, A_kp=A_kp)
    #     print(A_kp)
        f = f_j(img,kern).reshape(1,-1)[0]
    #     print(f.shape)
        X[i-1]=f

    A_kp=None
    for i in range(1,size+1):
        img_fn = '../project/Negative/'+("%05d" % i) +'.jpg'
        img = cv2.imread(img_fn,0)
        img, A_kp = normalize_grayscale(img, beta=0.75, image_id_k = i, A_kp=A_kp)
        f = f_j(img,kern).reshape(1,-1)[0]
        X[i+size-1]=f

    y=np.append(np.ones(size),np.zeros(size),axis=0)
    plt.show()
    
    avg_X = np.average(X,axis=0)
    X=np.subtract(X, avg_X)
    
    reg.fit(X,y)
    return reg
```

```python
def test(reg, test_size=120, train_size=200):
    A_kp=None
    X=np.zeros([test_size*2,227*227])
    for i in range(1+train_size,test_size+1+train_size):
        img_fn = '../project/Positive/'+("%05d" % i) +'.jpg'
        img = cv2.imread(img_fn,0)
        img, A_kp = normalize_grayscale(img, beta=0.75, image_id_k = i-size, A_kp=A_kp)
    #     print(A_kp)
        f = f_j(img).reshape(1,-1)[0]
    #     print(f.shape)
        X[i-1-train_size]=f

    A_kp=None
    for i in range(1+train_size,test_size+1+train_size):
        img_fn = '../project/Negative/'+("%05d" % i) +'.jpg'
        img = cv2.imread(img_fn,0)
        img, A_kp = normalize_grayscale(img, beta=0.75, image_id_k = i-size, A_kp=A_kp)
        f = f_j(img).reshape(1,-1)[0]
        X[i+test_size-train_size-1]=f


    y_pred = reg.predict(X)
    y_pred[:] = (np.sign(y_pred[:]-0.5)*1)
    return y_pred
```

```python
def check_error(reg, train_size, kern, image_offset=0):
    A_kp=None
    X=np.zeros([train_size*2,227*227])
    for i in range(1+image_offset,train_size+image_offset+1):
        img_fn = '../project/Positive/'+("%05d" % i) +'.jpg'
        img = cv2.imread(img_fn,0)
        img, A_kp = normalize_grayscale(img, beta=0.75, image_id_k = i-image_offset, A_kp=A_kp)
    #     print(A_kp)
        f = f_j(img,kern).reshape(1,-1)[0]
    #     print(f.shape)
        X[i-image_offset-1]=f

    A_kp=None
    for i in range(1+image_offset,train_size+image_offset+1):
        img_fn = '../project/Negative/'+("%05d" % i) +'.jpg'
        img = cv2.imread(img_fn,0)
        img, A_kp = normalize_grayscale(img, beta=0.75, image_id_k = i-image_offset, A_kp=A_kp)
        f = f_j(img,kern).reshape(1,-1)[0]
        X[i-image_offset+train_size-1]=f

    avg_X = np.average(X,axis=0)
    X=np.subtract(X, avg_X)
    
    y_pred = reg.predict(X)
    y_pred[:] = (np.sign(y_pred[:]-0.5)*0.5+0.5)
    return y_pred
```

```python
filter1 = naive_gabor(ksize=31, sigma=4, theta=np.pi/4, lamb=10, gamma=0.5, psi=0, ktype=cv2.CV_32F)
filter2 = naive_gabor(ksize=31, sigma=4, theta=np.pi/2, lamb=12, gamma=0.5, psi=0, ktype=cv2.CV_32F)
filter3 = naive_gabor(ksize=31, sigma=4, theta=3*np.pi/4, lamb=14, gamma=0.5, psi=0, ktype=cv2.CV_32F)
filter4 = naive_gabor(ksize=31, sigma=4, theta=0, lamb=16, gamma=0.5, psi=0, ktype=cv2.CV_32F)
reg1=train_weak_classifier_for_certain_filter(filter1)
reg2=train_weak_classifier_for_certain_filter(filter2)
reg3=train_weak_classifier_for_certain_filter(filter3)
reg4=train_weak_classifier_for_certain_filter(filter4)
reg_list = [reg1,reg2,reg3,reg4]
```

```python
def check_corr(a,b):
    if a==b:
        return 1
    else:
        return 0
    
vfunc_check_corr = np.vectorize(check_corr)
```

```python
def mult(beta,e):
    return beta**(1-e)
vfunc_mult = np.vectorize(mult)
vfunc_mult(4,[1,2,3])
```

```python
filter_list = [filter1,filter2,filter3,filter4]
```

```python
train_size = 200
weight=np.append(np.ones(train_size)/200,np.ones(train_size)/200,axis=0)
time=4
alpha_list = np.zeros(time)
H_list =[0]*time
max_idx_list = np.zeros(time)
for t in range(time):
    print('-----ITER {}-----------'.format(t))
    weight = normalize(weight)
    # for each feature:
    error_list = [0,0,0,0]
    y_pred_list = np.zeros((4,2*train_size))
    y_corr =np.append(np.ones(train_size),np.zeros(train_size),axis=0)
    for j in range(4):
        reg = reg_list[j]
        y_pred = check_error(reg,train_size, filter_list[j], image_offset=200)
        y_pred_list[j] = y_pred    
        print('y_predddddddd', y_pred)
        print('weighttttttttttttttt', weight)
        error_j = np.sum(np.multiply(np.absolute(y_pred-y_corr), weight))
        print('error is {} for feature {}'.format(error_j,j))
        error_list[j] = error_j
    # choose h with lowest error
    min_e = min(error_list)
    idx = np.argmin(error_list)
    max_idx_list[t] = idx
    beta= min_e/(1-min_e)
    print('beta {}\n  {} th feature with min error {}'.format(beta, idx, min_e))
    ei = vfunc_check_corr(y_pred_list[idx], y_corr)
    weight = vfunc_mult(beta, ei)
    alpha_t = np.log(1/beta)
    alpha_list[t] = alpha_t
    H_list[t]=reg_list[idx]
```

```python
def final_classification(img, H_list, max_idx_list, filter_list,alpha_list):
    final_prediction = 0
    for i in range(time):
        classifier_reg = H_list[i]
        idx=max_idx_list[i] 
        kern = filter_list[int(idx)]
        f = f_j(img,kern).reshape(1,-1)[0]
        pred_value = classifier_reg.predict(f)
        result = np.sign(pred_value-0.5)*1
        final_prediction += alpha_list[i]*result

    print("final_prediction: {}".format(final_prediction))
    print("threshold is {}".format(1/2*(np.sum(alpha_list))))
    
    returned_ans = 1
    if final_prediction > 1/2*(np.sum(alpha_list)):
        print("result is positive")
    else:
        print("negative")
        returned_ans = 0
    return returned_ans
```

```python
# Final strong classifier
test_size = 500
result = np.append(np.ones(test_size),np.zeros(test_size))
for i in range(1,test_size+1):
    img_fn = '../project/Positive/'+("%05d" % i) +'.jpg'
    img = cv2.imread(img_fn,0)
    img, A_kp = normalize_grayscale(img, beta=0.75, image_id_k = i, A_kp=A_kp)
    print('========={}========'.format(i))
#     print(final_classification(img, H_list, max_idx_list, filter_list, alpha_list))
    result[i-1]=final_classification(img, H_list, max_idx_list, filter_list, alpha_list)
    
for i in range(1,test_size+1):
    img_fn = '../project/Negative/'+("%05d" % i) +'.jpg'
    img = cv2.imread(img_fn,0)
    img, A_kp = normalize_grayscale(img, beta=0.75, image_id_k = i, A_kp=A_kp)
    result[i+test_size-1]=final_classification(img, H_list, max_idx_list, filter_list, alpha_list)

print(result)
```

```python
y1=result[0:test_size]
y2=result[test_size:]
print("positive ", y1)
unique, counts = np.unique(y1, return_counts=True)
r1= dict(zip(unique, counts))
print(r1)
TP = r1[1]/float(test_size)
FN = r1[0]/float(test_size)
print("TP: {}, FN: {}".format(TP, FN))

print("\n negative")
unique, counts = np.unique(y2, return_counts=True)
r2=dict(zip(unique, counts))
TN = r2[0]/float(test_size)
FP = r2[1]/float(test_size)
print("TN: {}, FP: {}".format(TN, FP))

prec = TP/(TP+FP)
recall = TP/(TP+FN)
print("\n precision:{} , recall: {}".format(prec, recall))
```
